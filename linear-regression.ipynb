# Split the Data

```python
x = z['x'].values
```

```python
y =z['y'].values
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y)
```

```python
x_train = x_train.reshape(-1,1)
x_test = x_test.reshape(-1,1)
```

```python
model = LinearRegression()
```

# Train the model

```python
model.fit(x_train,y_train)
```


### Linear Regression

The formula for the **Linear Regression** model is:

$$
y = \beta_0 + \beta_1 x
$$

where:

- \( y \) is the dependent variable (target value).
- \( x \) is the independent variable (feature).
- \( \beta_0 \) is the y-intercept.
- \( \beta_1 \) is the slope of the line.

To calculate the slope \( \beta_1 \) and intercept \( \beta_0 \), we use the following formulas:

$$
\beta_1 = \frac{\sum_{i=1}^{n} (x_i - \bar{x}) \times (y_i - \bar{y})}{\sum_{i=1}^{n} (x_i - \bar{x})^2}
$$

$$
\beta_0 = \bar{y} - \beta_1 \bar{x}
$$

where:

- (bar x ) is the mean of the independent variable \( x \).
-  (bar y ) is the mean of the dependent variable \( y \).
- ( n ) is the number of data points.


```python
print(f"Interept Is {model.intercept_}")
print(f"Coificent is{model.coef_}")
```

```python
y_pred = model.predict(x_test)
```

# accuracy

```python
mse = mean_squared_error(y_test , y_pred)
```

```python
print(f"Mean squared Error {mse}")
```

The formula for Mean Squared Error (MSE) is:

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

where:

- \( y_i \) is the actual value.
- \( \hat{y}_i \) is the predicted value.
- \( n \) is the number of data points.


```python
r2 = r2_score(y_test,y_pred)
```

```python
print(f"r2 Score is{r2}")
```

The formula for \( R^2 \) is:

$$
R^2 = 1 - \frac{\text{SS}_{\text{res}}}{\text{SS}_{\text{tot}}}
$$



where:

- \( \text{SS}_{\text{res}} \) is the Residual Sum of Squares:
$$
\text{SS}_{\text{res}} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$



- \( \text{SS}_{\text{tot}} \) is the Total Sum of Squares:
$$
\text{SS}_{\text{tot}} = \sum_{i=1}^{n} (y_i - \bar{y})^2
$$



Here:

- \( y_i \) is the actual value.
- \( \hat{y}_i \) is the predicted value.
- \( \bar{y} \) is the mean of the actual values.
- \( n \) is the number of data points.
